```python
import pandas as pd
from collections import Counter
```

```python
# read file
df_13 = pd.read_csv('pub_pri_mfcc_13.csv', header= None)
df_17 = pd.read_csv('pub_pri_mfcc_17.csv', header= None)
df_21 = pd.read_csv('pub_pri_mfcc_21.csv', header= None)
df_30 = pd.read_csv('pub_pri_mfcc_30.csv', header= None)
df_50 = pd.read_csv('pub_pri_mfcc_50.csv', header= None)
list_all = list(zip(df_13[1].to_list(), df_17[1].to_list(), df_21[1].to_list(), df_30[1].to_list(), df_50[1].to_list()))
```

```python
# ensemble
outputs = []
for a, ii in enumerate(list_all):
    class_counts = Counter(ii)
    most_common_classes = class_counts.most_common()

    if len(most_common_classes) > 1 and most_common_classes[0][1] == most_common_classes[1][1]:
        predicted_class = max(ii)
    else:
        predicted_class = most_common_classes[0][0]  # 否则选择出现次数最多的类别作为最终预测结果

    outputs.append(predicted_class)
```

```python
tem_pub_pri = pd.read_csv(r'..\Private Testing Dataset\submission_template_public+private.csv',header=None)
tem_pub_pri[1] = outputs
tem_pub_pri.to_csv('ensemble.csv', header=False, index=False)
```
